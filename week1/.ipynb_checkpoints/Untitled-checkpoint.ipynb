```python
import pandas as pd
import csv
import json
import random
df = pd.read_csv('./international.csv')
df
```

```python
asn = df.ASN.unique()
# asn
```

```python
nodes = []
for i in asn:
    r = lambda: random.randint(0,255)
    color = '#%02X%02X%02X' % (r(),r(),r())
    s = random.uniform(0,100)
    node = {
        "color": color,
        "label": i,
        "x": 4.2,
        "y": 4.2,
        "attributes": {},
        "id": i,
        "size": s
    }
    nodes.append(node)
edges = []
for index, row in df.iterrows():
    edge = {
        "sourceID": row['ASN-Source'],
        "attributes": {},
        "targetID": row['ASN'],
        "size": 1
    }
    edges.append(edge)
filejson = json.dumps({'nodes': nodes, 'edges': edges}, indent=4, separators=(',', ': '))
print(nodes)
```
